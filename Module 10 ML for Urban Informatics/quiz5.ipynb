# Air BnB

```python
air_bnb_csv = 'Quiz5_Air_BnB_in_NYC.csv'
air_bnb_df = pd.read_csv(air_bnb_csv)
air_bnb_df
```

```python
air_bnb_df.info()
```

```python
air_bnb_df['room_type'].value_counts()/len(air_bnb_df)
```

```python
boroughs = ['Bronx', 'Brooklyn', 'Manhattan', 'Queens', 'Staten Island']
borough_properties = air_bnb_df[air_bnb_df['neighbourhood_group'].isin(boroughs)]

borough_properties['neighbourhood_group'].value_counts()/len(borough_properties)
```

```python
top_20_reviewed = air_bnb_df.sort_values(by='number_of_reviews', ascending=False)[:20]
top_20_reviewed
```

```python
top_20_reviewed.groupby(by='neighbourhood')['number_of_reviews'].mean().sort_values(ascending=False)
```

```python
manhattan_properties = air_bnb_df[air_bnb_df['neighbourhood_group'] == 'Manhattan']
manhattan_properties['neighbourhood'].value_counts(ascending=False)
```

```python
upper_east_properties = manhattan_properties[manhattan_properties['neighbourhood'] == 'Upper East Side']
upper_east_properties
```

```python
percentile_85 = np.quantile(upper_east_properties['number_of_reviews'], 0.85)
upper_east_85_percentile = upper_east_properties[upper_east_properties['number_of_reviews'] >= percentile_85]
```

```python
upper_east_private_rooms = upper_east_85_percentile[upper_east_85_percentile['room_type'] == 'Private room']
cheapest_upper_east_pr = upper_east_private_rooms.loc[upper_east_private_rooms['price'].idxmin()]
cheapest_upper_east_pr
```

```python
upper_east_entire = upper_east_85_percentile[upper_east_85_percentile['room_type'] == 'Entire home/apt']
cheapest_upper_east_entire = upper_east_entire.loc[upper_east_entire['price'].idxmin()]
cheapest_upper_east_entire
```

```python
price_gap = np.abs(cheapest_upper_east_pr['price'] - cheapest_upper_east_entire['price'])
price_gap
```

# Uber

```python
uber_df = pd.read_csv('UberData.csv')
uber_df
```

```python
uber_df.info()
```

```python
dt_columns = [
    'Request Time',
    'Begin Trip Time',
    'Dropoff Time'
]

for dt in dt_columns:
    uber_df[dt] = pd.to_datetime(uber_df[dt].str.replace(r' \+0000 UTC', '', regex=True))

uber_df.info()
```

```python
uber_df
```

```python
uber_df['Product Type'].value_counts()
```

```python
jan_14 = uber_df[
    (uber_df['Product Type'].fillna('').str.contains('uberx', case=False)) &
    (uber_df['Trip or Order Status'] == 'COMPLETED') &
    (uber_df['Begin Trip Time'].dt.date == pd.to_datetime('2020-01-14').date())
]

jan_14
```

```python
uber_df['request_hour'] = uber_df['Request Time'].dt.hour
uber_df['request_hour'].value_counts()
```

```python
uber_df['request_dow'] = uber_df['Request Time'].dt.day_of_week
uber_df['request_dow'].value_counts()
```
